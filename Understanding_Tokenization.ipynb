```python
import tiktoken
```

```python
encoding = tiktoken.encoding_for_model("gpt-4.1-mini")
```

```python
tokens = encoding.encode("Hello, my name is Ian!")
```

```python
tokens
```

```python
for id in tokens:
  text = encoding.decode([id])
  print(f"{id} is {text}")
```

```python
from google.colab import userdata
userdata.get('OPENAI_API_KEY')
userdata.get('Gemini')
```

```python
from openai import OpenAI
openai = OpenAI(
    api_key=userdata.get('OPENAI_API_KEY')
)
```

```python
gemini = OpenAI(
    api_key=userdata.get('Gemini'),
    base_url="https://generativelanguage.googleapis.com/v1beta"
)
```

```python
gemini_say = "Hi, GPT! I am Gemini from Google"
gpt_say = "Hi, Gemini!, I am Chatgpt5"
```

```python
gpt_message = [
    {"role": "system", "content": "you are very snarky and argumentative"},
    {"role": "user", "content": "Hi, GPT! I am Gemini from Google"}
]
```

```python
gemini_message = [
    {"role": "system", "content": "you are very smart and kind"},
    {"role": "user", "content": "Hi, Gemini!, I amChatGPT5"}
]
```

```python
gemini_model = "gemini-2.5-flash"
gpt_model = "gpt-4.1-mini"
```

```python
gpt_response = openai.chat.completions.create(
    model=gpt_model,
    messages=gpt_message)
```

```python
gemini_response = gemini.chat.completions.create(
    model=gemini_model,
    messages=gemini_message,
)
```

Introduction -----

```python
gpt_say = gpt_response.choices[0].message.content
```

```python
gemini_say = gemini_response.choices[0].message.content
```

```python
gpt_message = [
    {"role": "system", "content": "you are someone that is very snarky and argumentative"},
    {"role": "assistant", "content": f"{gpt_say}"},
    {"role": "user", "content": f"{gemini_say}"}
     ]
```

```python
gemini_message = [
    {"role": "system", "content": "you are very polite and kind person"},
    {"role": "assistant", "content": f"{gemini_say}"},
    {"role": "user", "content": f"{gpt_say}"}
     ]
```

```python
def gpt_reply():
  gpt_response = openai.chat.completions.create(
      model=gpt_model,
      messages=gpt_message)
  reply = gpt_response.choices[0].message.content
  gpt_say = reply
  print(reply)
```

```python
def gemini_reply():
  gemini_response = gemini.chat.completions.create(
      model=gemini_model,
      messages=gemini_message)
  reply = gemini_response.choices[0].message.content
  gemini_say = reply
  print(f"{reply}\n\n")
```

```python
def intro():
  print("GEMINI:")
  print(f"{gemini_say}\n\n")
  print("GPT:")
  print(f"{gpt_say}\n")
```

```python
def conversation():
  print("GEMINI:")
  gemini_reply()
  print("\n")
  print("GPT:")
  gpt_reply()
  print("\n")
```

```python
intro()
for iantristan in range(6):
  conversation()
```

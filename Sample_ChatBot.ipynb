```python
from gradio.events import api
import gradio as gr
from openai import OpenAI
from google.colab import userdata
userdata.get('OPENAI_API_KEY')
openai = OpenAI(api_key=userdata.get('OPENAI_API_KEY'))
```

```python
def stream_output(prompt):
    messages = [
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": prompt}
    ]

    stream = openai.chat.completions.create(
        model="gpt-4o-mini",   # choose a model
        messages=messages,
        stream=True
    )

    result = ""
    for chunk in stream:
        delta = chunk.choices[0].delta.content
        if delta:
            result += delta
            yield result
```

```python
Dark_mode = """
function refresh() {
  const url = new URL(window.location);
  if (url.searchParams.get('__theme') !== 'dark') {
    url.searchParams.set('__theme', 'dark');
    window.location.href = url.href;
  }
}
"""
```

```python
message_input = gr.Textbox(label="Ask Anything",info="Ask Anything",lines=7)

message_output = gr.Markdown(label="Response:")

view = gr.Interface(
    fn=stream_output,
    title="Donkee's AI by Ian Tristan Cultura",
    inputs=message_input,
    outputs=message_output,
    examples=[
        "ask anything",
        "hello",
        "what can you do?",
        "explain AI"
    ],
    flagging_mode="never",
    js=Dark_mode
)

view.launch()
```
